```python
import pandas as pd
import numpy as np
```

**Groupby**

```python
data = {
    "category": [
        "Electronics", "Electronics", "Clothing", "Clothing",
        "Grocery", "Grocery", "Electronics", "Clothing",
        "Grocery", "Electronics", "Clothing", "Grocery"
    ],
    "store": [
        "Store A", "Store B", "Store A", "Store C",
        "Store B", "Store A", "Store A", "Store B",
        "Store C", "Store C", "Store A", "Store B"
    ],
    "sales": [
        1200, 800, 150, 300,
        90, 60, 600, 200,
        120, 1500, 100, 75
    ],
    "quantity": [
        2, 1, 3, 5,
        10, 8, 1, 4,
        12, 3, 2, 7
    ],
    "date": [
        "2024-01-01", "2024-01-01", "2024-01-01", "2024-01-02",
        "2024-01-02", "2024-01-02", "2024-01-03", "2024-01-03",
        "2024-01-03", "2024-01-04", "2024-01-04", "2024-01-04"
    ]
}

df= pd.DataFrame(data)
df
```

```python
# Group by category and sum sales
df.groupby("category")["sales"].sum()
```

```python
# Group by store and sum sales
df.groupby("store")["sales"].sum()
```

```python
#group by multiple columns 
#Group by store and category and sum sales
df.groupby(["category", "store"])["sales"].sum()
```

**Aggregation**

```python
# Calculate the average (mean) of sales
df["sales"].mean()

# Find the maximum sales value
df["sales"].max()

# Find the minimum sales value
df["sales"].min()

# Calculate the total sales
df["sales"].sum()

# Count the number of sales records
df["sales"].count()
```
