A Pandas Series is a one-dimensional (1D) labeled data structure provided by the Pandas library in Python.

In simple terms:
Data + Index (labels) = Pandas Series

CREATING PANDAS SERIES

There are multiple ways to create pandas series

```python
import numpy as np
import pandas as pd
```

```python
labels=['a','b','c']
my_list=[1,2,3]
arr= np.array([2,5,7])
d={1:12,2:22,3:33}
```

```python
#creating series with list
pd.Series(my_list)
```

```python
#giving lables to the list
pd.Series(my_list,index=lables)
```

```python
#Creating Series with numpy array
pd.Series(arr)
```

```python
#here also we can apply labels 
pd.Series(arr,index=lables)
```

```python
#We can also create pandas series using dictionary but here we cannot add labels
pd.Series(d)
```
