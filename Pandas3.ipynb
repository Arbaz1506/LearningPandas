```python
import pandas as pd
import numpy as np
```

Finding missing Data

```python

data = {
    "A": [10, 20, np.nan, 40],
    "B": [np.nan, 25, 35, 45],
    "C": [15, np.nan, 30, np.nan],
    "D": [np.nan, 22, 32, 42]
}


df = pd.DataFrame(data)
df
```

```python
#finding nan value
df.isna()
```

```python
#To find total Nan value
df.isna().sum()
```

```python
# Checks for NaN values in the DataFrame
# Returns True for columns that contain at least one NaN value
df.isna().any()
```

```python
# Checks for NaN values row-wise
# Returns True for rows that contain at least one NaN value
df.isna().any(axis=1)
```
