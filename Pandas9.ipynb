```python
# Select the row(s) corresponding to the country with the highest population
df[df['population'] == df['population'].max()]['country']
```

```python
# Select the capital city of the country that has the highest population
df[df['population'] == df['population'].max()]['capital_city']
```

```python
# Select the row(s) corresponding to the country with the least population
df[df['population'] == df['population'].min()]['country']
```

```python
# Select the capital city of the country that has the lowest population
df[df['population'] == df['population'].min()]['capital_city']
```

```python
#List top five country with highest democratic value
df.sort_values(by='democracy_score', ascending=False,inplace=True)
df.head()
```

```python
#How many total regions are there
df['region'].value_counts().count()
```

```python
#How many countries lies in eastern region
df['region'].value_counts()['Eastern Europe']
```

```python
# Who is the political leader of the 2nd highest populated country
df[df['population'] == df['population'].nlargest(2).iloc[1]]['political_leader']
```

```python
# How many countries are there whose political leaders are unknown
df['political_leader'].isna().sum()
```
